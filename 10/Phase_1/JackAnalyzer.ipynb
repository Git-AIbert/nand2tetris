```python
import import_ipynb
import os.path
from JackTokenizer import JackTokenizer
```

```python
path = '/Users/ceshi/Desktop/计算机系统要素/nand2tetris/projects/10/Square/Main'
fileIn = open(path + '.jack', 'r')
fileOut = open(path + 'TT.xml', 'w')
jackTokenizer = JackTokenizer(fileIn)
outString = '<tokens>\n'
while(jackTokenizer.hasMoreTokens()):
    jackTokenizer.advance()
    if jackTokenizer.tokenType() == 'KEYWORD':
        outString += '<keyword> ' + jackTokenizer.keyword() + ' </keyword>\n'
    if jackTokenizer.tokenType() == 'SYMBOL':
        outString += '<symbol> ' + jackTokenizer.symbol() + ' </symbol>\n'
    if jackTokenizer.tokenType() == 'IDENTIFIER':
        outString += '<identifier> ' + jackTokenizer.identifier() + ' </identifier>\n'
    if jackTokenizer.tokenType() == 'INT_CONST':
        outString += '<integerConstant> ' + str(jackTokenizer.intVal()) + ' </integerConstant>\n'
    if jackTokenizer.tokenType() == 'STRING_CONST':
        outString += '<stringConstant> ' + jackTokenizer.stringVal() + ' </stringConstant>\n'
outString += '</tokens>'
print(outString)
fileOut.write(outString)
fileIn.close()
fileOut.close()
```
