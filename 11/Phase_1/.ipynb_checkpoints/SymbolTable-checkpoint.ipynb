```python
class SymbolTable:
    
    def __init__(self):
        self.staticIndex = 0
        self.fieldIndex = 0
        self.argumentIndex = 0
        self.varIndex = 0
        self.classScope = {}
        self.subroutineScope = {}
        return
    
    def startSubroutine(self):
        self.fieldIndex = 0
        self.argumentIndex = 0
        self.varIndex = 0
        self.subroutineScope.clear()
        return
    
    def Define(self, Name, Type, Kind):
        if Kind == 'static':
            self.classScope[Name] = [Type, Kind, self.staticIndex]
            self.staticIndex += 1
        if Kind == 'field':
            self.classScope[Name] = [Type, Kind, self.fieldIndex]
            self.fieldIndex += 1
        if Kind == 'argument':
            self.subroutineScope[Name] = [Type, Kind, self.argumentIndex]
            self.argumentIndex += 1
        if Kind == 'var':
            self.subroutineScope[Name] = [Type, Kind, self.varIndex]
            self.varIndex += 1
        return
    
    def VarCount(self, Kind):
        if Kind == 'static':
            return staticIndex
        if Kind == 'field':
            return self.fieldIndex
        if Kind == 'argument':
            return self.argumentIndex
        if Kind == 'var':
            return self.varIndex
    
    def KindOf(self, Name):
        if Name in self.subroutineScope:
            return self.subroutineScope[Name][1]
        elif Name in self.classScope:
            return self.classScope[Name][1]
        else:
            return 'None'
    
    def TypeOf(self, Name):
        if Name in self.subroutineScope:
            return self.subroutineScope[Name][0]
        elif Name in self.classScope:
            return self.classScope[Name][0]
        else:
            return 'None'
    
    def IndexOf(self, Name):
        if Name in self.subroutineScope:
            return self.subroutineScope[Name][2]
        elif Name in self.classScope:
            return self.classScope[Name][2]
        else:
            return 'None'
```
